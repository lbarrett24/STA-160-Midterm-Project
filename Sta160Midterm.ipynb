```python
plt.figure(figsize=(15,40))
for i,column in enumerate(factor):
    plt.subplot(len(factor), 2, i+1)
    plt.suptitle("Plot Value Count", fontsize=20, x=0.5, y=1)
    sns.countplot(data=df, x=column)
    plt.title(f"{column}")
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,50))
for i,column in enumerate(factor[1:]):
    plt.subplot(len(factor), 2, i+1)
    plt.suptitle("Plot Value Count VS HeartAttack", fontsize=20, x=0.5, y=1)
    sns.countplot(data=df, x=column, hue='HeartDiseaseorAttack')
    plt.title(f"{column}")
    plt.tight_layout()
```

```python
import pandas as pd
from scipy.stats import entropy

# Assume the DataFrame is named df
entropies = []
for col in df.columns:
    if df[col].dtype == 'category':
        ent = entropy(df[col].value_counts(normalize=True), base=2)
        entropies.append((col, ent))

# Sort the list of entropies by descending order
entropies = sorted(entropies, key=lambda x: x[1], reverse=True)

# Print the list of entropies and column names
for col, ent in entropies:
    print(col, ent)
```
