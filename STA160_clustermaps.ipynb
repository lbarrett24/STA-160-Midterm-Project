```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Read in cleaned data set
df = pd.read_csv("/Users/calvinatkins/Downloads/Data Sets/HD_indicators_cleaned.csv", index_col = 0)
df.astype('category')
df.dtypes
```

```python
# select data of interest (Example)
# not neccesarily needed for this
data = df[['HeartDiseaseorAttack', 'HighBP', 'Sex']]
```

```python
# create contingency tables with count
contingency_table = pd.crosstab(df['HighBP'], df['HeartDiseaseorAttack'])

# create contingency table with conditonal probability on row sums
contingency_table = pd.crosstab(df['HighBP'], df['HeartDiseaseorAttack'], normalize='index')
print(contingency_table)

# create clustermap
sns.clustermap(contingency_table, cmap = "coolwarm")
```

```python
# create contingency tables with count
contingency_table = pd.crosstab(df['HighBP'], [df['HeartDiseaseorAttack'], df['Stroke']])

# create contingency table with conditonal probability on row sums
contingency_table = pd.crosstab(df['HighBP'], [df['HeartDiseaseorAttack'], df['Stroke']], normalize='index')
print(contingency_table)

# create clustermap
sns.clustermap(contingency_table, cmap = "coolwarm")
```

```python
# create contingency table with conditonal probability on row sums
contingency_table = pd.crosstab([df['Sex'],df['HighBP']], [df['HeartDiseaseorAttack'], df['Stroke']], normalize='index')
print(contingency_table)

# create clustermap
sns.clustermap(contingency_table, cmap = "coolwarm")
```
