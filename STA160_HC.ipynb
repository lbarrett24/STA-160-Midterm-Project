```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in cleaned data set
df = pd.read_csv("/Users/calvinatkins/Downloads/Data Sets/HD_indicators_cleaned.csv", index_col = 0)
df
```

```python
# selected data
# for learning about this, I will just choose, HighBP, HighChol, and Sex

selected_data = df[['HeartDiseaseorAttack', 'HighBP', 'Age']]
selected_data
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(selected_data)
data_scaled = pd.DataFrame(data_scaled, columns=selected_data.columns)
data_scaled
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='average'))
```
